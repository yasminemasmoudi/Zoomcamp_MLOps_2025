Now let's compute the duration variable. It should contain the duration of a ride in minutes.

What's the standard deviation of the trips duration in January?

```python
# Compute duration in minutes
yellow_tripdata_january['duration'] = (yellow_tripdata_january['tpep_dropoff_datetime'] - yellow_tripdata_january['tpep_pickup_datetime']).dt.total_seconds() / 60
yellow_tripdata_january.head(5)
```

```python
# Compute standard deviation
std_duration = yellow_tripdata_january['duration'].std()
print(round(std_duration, 2))
```

Next, we need to check the distribution of the duration variable. There are some outliers. Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive).

What fraction of the records left after you dropped the outliers?

To compute the fraction of records remaining after filtering out outliers, we need to:

1. Calculate trip duration in minutes.

2. Filter to keep only trips with duration between 1 and 60 minutes inclusive.

3. Divide the number of filtered records by the total number of records.

```python
# Total number of records before filtering
total_records = len(yellow_tripdata_january)
print(total_records)
```

```python
# Filter durations between 1 and 60 minutes inclusive
filtered_yellow_tripdata_january = yellow_tripdata_january[(yellow_tripdata_january['duration'] >= 1) & (yellow_tripdata_january['duration'] <= 60)]
filtered_yellow_tripdata_january.head(5)
```

```python
# Number of records after filtering
filtered_records = len(filtered_yellow_tripdata_january)
print(filtered_records)
```

```python
# Fraction remaining
fraction_remaining = filtered_records / total_records
print(round(fraction_remaining * 100, 2))
```

Let's apply one-hot encoding to the pickup and dropoff location IDs. We'll use only these two features for our model.

Turn the dataframe into a list of dictionaries (remember to re-cast the ids to strings - otherwise it will label encode them)
Fit a dictionary vectorizer
Get a feature matrix from it
What's the dimensionality of this matrix (number of columns)?

```python
yellow_tripdata_january.head(5)
```

We're now preparing data for modeling using one-hot encoding with a DictVectorizer, focusing only on PULocationID and DOLocationID

```python
# Filter durations between 1 and 60 mins
yellow_tripdata_january = yellow_tripdata_january[(yellow_tripdata_january['duration'] >= 1) & (yellow_tripdata_january['duration'] <= 60)]
```

```python
# Convert location IDs to string (to one-hot encode correctly)
yellow_tripdata_january['PULocationID'] = yellow_tripdata_january['PULocationID'].astype(str)
yellow_tripdata_january['DOLocationID'] = yellow_tripdata_january['DOLocationID'].astype(str)
```

```python
# Create list of dictionaries for DictVectorizer
dicts = yellow_tripdata_january[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer

# Fit DictVectorizer
dv = DictVectorizer()
X = dv.fit_transform(dicts)

# Get the shape (rows, columns)
print(X.shape)
```

Now let's use the feature matrix from the previous step to train a model.

Train a plain linear regression model with default parameters, where duration is the response variable.

Calculate the RMSE of the model on the training data.

What's the RMSE on train?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Train the model
model = LinearRegression()
model.fit(X, yellow_tripdata_january['duration'])

# Predict on training data
y_pred = model.predict(X)

# Calculate RMSE
mse = mean_squared_error(yellow_tripdata_january['duration'], y_pred)
rmse = np.sqrt(mse)
print(rmse)
```

Now let's apply this model to the validation dataset (February 2023).

What's the RMSE on validation?

```python
# Load February data
df_val = pd.read_parquet('./data/yellow_tripdata_2023-02.parquet')
```

```python
# Compute duration
df_val['duration'] = (df_val['tpep_dropoff_datetime'] - df_val['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filter duration between 1 and 60 mins
df_val = df_val[(df_val['duration'] >= 1) & (df_val['duration'] <= 60)]
```

```python
# Convert IDs to string
df_val['PULocationID'] = df_val['PULocationID'].astype(str)
df_val['DOLocationID'] = df_val['DOLocationID'].astype(str)

# Create dicts and transform using the SAME vectorizer
val_dicts = df_val[['PULocationID', 'DOLocationID']].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
# Predict on validation data
y_pred_val = model.predict(X_val)

# Calculate MSE and then RMSE manually
mse_val = mean_squared_error(df_val['duration'], y_pred_val)
rmse_val = np.sqrt(mse_val)
print(rmse_val)
```
